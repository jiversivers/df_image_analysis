```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
sigma = [1, 10]
theta = [np.pi/4]
lamba = [np.pi/4, np.pi/2, np.pi]
gamma = [1, 10]
phi = [np.pi/4, np.pi]
print(f'Total different kernels: {len(sigma) * len(theta) * len(lamba) * len(gamma) * len(phi)}')
```

```python
x = np.linspace(-5, 5, 9)  # X-coordinates
y = np.linspace(-5, 5, 9)  # Y-coordinates
X, Y = np.meshgrid(x, y)    # Create a 2D grid
fig, axes = plt.subplots(6,6, figsize = (36,36), subplot_kw={'projection': '3d'})
i = 0

for s in sigma: 
    for t in theta: 
        for l in lamba:
            for g in gamma:
                for p in phi:
                    
                    kernel = cv.getGaborKernel((9, 9), s, t, l, g, p)
                    r = i // 6
                    c = i % 6
                    ax = axes[r, c]
                    ax.set_title(f's:{s:0.2f}, t:{t:0.2f}, l:{l:0.2f}, g:{g:0.2f}, p:{p:0.2f}')
                    ax.plot_surface(X, Y, kernel, rstride=1, cstride=1, cmap='viridis')
                    i += 1
                    
```
